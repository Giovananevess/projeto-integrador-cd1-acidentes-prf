```python
import pandas as pd

df_2024 = pd.read_csv(
    "../data/raw/acidentes_prf_2024.csv",
    sep=";",
    encoding="latin1"
)

df_2025 = pd.read_csv(
    "../data/raw/acidentes_prf_2025.csv",
    sep=";",
    encoding="latin1"
)

df = pd.concat([df_2024, df_2025], ignore_index=True)

print(df.shape)
df.head()
```

```python
# Informações gerais
print("Total de acidentes:", len(df))
print("Total de UFs:", df["uf"].nunique())
print("Total de mortos:", df["mortos"].sum())
print("Total de feridos:", df["feridos"].sum())
```
